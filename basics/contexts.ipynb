```python
class enclosing_tag:
    def __init__(self, tagname):
        self.tagname = tagname

    def __enter__(self):
        print(f'<{self.tagname}>', end='')

    def __exit__(self, etyp, einst, etb):
        print(f'</{self.tagname}>')


with enclosing_tag('sometag'):
    pass
```

```python
import contextlib

@contextlib.contextmanager
def enclosing_tag(tagname):
    print(f'<{tagname}>', end='')

    try:
        yield
    finally:
        print(f'</{tagname}>')


with enclosing_tag('sometag'):
    pass
```
