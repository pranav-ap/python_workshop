# Closure

## Definition

A closure is a function that:

- Is defined inside another function,
- Captures variables from the outer function’s scope,
- And remembers those variables even after the outer function has finished executing.

```python
def make_adder(n): # outer function
    def add(x):  # closure
        return x + n

    return add
```

```python
plus_3 = make_adder(3)
plus_3(10)
```

## Callable Objects

```python
class Adder:
    def __init__(self, n):
        self.n = n
    def __call__(self, x):
        return self.n + x
```

```python
plus_3 = Adder(3)
plus_3(10)
```

```python
callable(plus_3)
```

## nonlocal

```python
# def outer():
#     x = 10
#     def inner():
#         x += 1  # ❌ UnboundLocalError
#         return x
#     inner()
```

nonlocal tells Python:

This variable is not local to the inner function —
go to the nearest enclosing function and use that variable.

```python
def outer():
    x = 10

    def inner():
        nonlocal x
        x += 1
        return x

    return inner


f = outer()
print(f())  # 11
print(f())  # 12
```

## global

```python
x = 20

def outer():
    x = 10

    def inner():
        global x
        x += 1
        return x

    return inner


f = outer()
print(f())  # 21
print(f())  # 22
```

# Decorator

```python
def simplest_possible_decorator(func):
    return func

@simplest_possible_decorator
def greet():
    return "Hello!"

greet()
```

```python
def simplest_possible_decorator(func):
    return func

def greet():
    return 'Hello!'

greet = simplest_possible_decorator(greet)

greet()
```

```python
def uppercase(func):
    def wrapper():
        original_result = func()
        modified_result = original_result.upper()
        return modified_result

    return wrapper

@uppercase
def greet():
    return "Hello!"

greet()
```

## Mulitple Decorators

```python
def emphasis(func):
    def wrapper():
        return '<em>' + func() + '</em>'
    return wrapper


# greet = emphasis(uppercase(greet))


@emphasis # applied second
@uppercase # applied first
def greet():
    return "Hello!"


greet()
```

## Forward args

```python
def trace(func):
    def wrapper(*args, **kwargs):
        print(f'TRACE: calling {func.__name__}() with {args}, {kwargs}')
        original_result = func(*args, **kwargs)
        print(f'TRACE: {func.__name__}() returned {original_result!r}')
        return original_result

    return wrapper


@trace
def say(name, line):
    return f'{name}: {line}'


say('Alice', 'Hello, World!')
```

# *args, **kwargs

## Packing

```python
def foo(required, *args, **kwargs):
    """
    requires at least one argument called “required,”
    but it can accept extra positional and keyword arguments as well.
    """
    print(required)
    if args:
        print(args)
    if kwargs:
        print(kwargs)
```

```python
foo('hello', 1, 2)
```

```python
foo('hello', 1, 2, 3, key1='value', key2=999)
```

## Unpacking

```python
def bar(x, *args, name='Unknown', age=0):
    print(f'x={x}, args={args}, name={name}, age={age}')


def foo(x, *args, **kwargs):
    kwargs['name'] = 'Alice'
    new_args = args + ('extra', )
    bar(x, *new_args, **kwargs)


foo(1, 2, 3, age=30)
```
